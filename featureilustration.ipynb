```python
import librosa
import matplotlib.pyplot as plt
import numpy as np

audio_file = 'fma_small/000/000002.mp3'  
y, sr = librosa.load(audio_file, sr=None)
y=y[:sr*10]
plt.figure(figsize=(10, 4))
plt.plot(np.linspace(0, len(y) / sr, len(y)), y)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
print(y.shape)
```

```python
# Chroma STFT
chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
plt.figure(figsize=(10, 4))
librosa.display.specshow(chroma_stft, x_axis='time', sr=sr)
plt.colorbar()
plt.ylabel('Chroma Bin')
plt.show()
```

```python
# Spectral Centroid
spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)[0]
plt.figure(figsize=(10, 4))
plt.semilogy(np.linspace(0, len(y) / sr, len(spectral_centroid)), spectral_centroid)
plt.xlabel('Time (s)')
plt.ylabel('Hz')
plt.show()
```

```python
# Spectral Contrast
spectral_contrast = librosa.feature.spectral_contrast(y=y, sr=sr)
plt.figure(figsize=(10, 4))
librosa.display.specshow(spectral_contrast, x_axis='time', sr=sr, cmap='coolwarm')
plt.colorbar()
plt.ylabel('Frequency Band')
plt.show()
```

```python
# Zero Crossing Rate
zero_crossing_rate = librosa.feature.zero_crossing_rate(y)[0]
plt.figure(figsize=(10, 4))
plt.plot(np.linspace(0, len(y) / sr, len(zero_crossing_rate)), zero_crossing_rate)
plt.xlabel('Time (s)')
plt.ylabel('Rate')
plt.show()
```

```python
# MFCCs
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time', sr=sr, cmap='coolwarm')
plt.colorbar()
plt.ylabel('MFCC Coefficient')
plt.show()
```
